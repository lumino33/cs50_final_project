```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
background = np.ones((1000,1000,3), np.uint8)*255.0

height_canon_image, width_canon_image = 100, 100
canon_image = cv2.imread("images/canon_icon.png", -1)
canon_image = cv2.resize(canon_image, (height_canon_image, width_canon_image), interpolation = cv2.INTER_AREA)
```

```python
background[1000-height_canon_image:, :width_canon_image, :] = canon_image[:,:,:3]
```

```python
plt.figure(figsize = (6, 10))
plt.imshow(background/255.0)
plt.show()
```
